```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x1,y1,x2,y2 = [],[],[],[]

#RED DATAPOINTS
for i in range(100):
    x1.append(random.randint(35,70))
for i in range(100):
    y1.append(random.randint(45,80))
    
#BLUE DATAPOINTS
for i in range(100):
    x2.append(random.randint(10,40))
for i in range(100):
    y2.append(random.randint(10,40))
    
    
#x1 = np.array(x1)
#x2= np.array(x2)
#y1 = np.array(x2)
#y2 = np.array(y2)
```

```python
plt.xlim(0,80)
plt.ylim(0,90)
plt.scatter(x1,y1,c='red')
plt.scatter(x2,y2,c='blue')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.datasets import load_iris
```

```python
l=[]
for i in range(len(x1)):
    l.append([x1[i],y1[i]])
for i in range(len(x1)):
    l.append([x2[i],y2[i]])
```

```python
k = []
for i in range(201):
    if i<100:    
        k.append(1)
    if i>100:
        k.append(0)
```

```python
arr = np.array(l)
```

```python
df =pd.DataFrame(arr,columns=['A','B'])
df['target'] = np.array(k)
```

```python
sns.scatterplot(x='A',y='B',data=df,hue='target')
```

```python
sns.pairplot(df,hue='target')
```

```python
X = df[['A','B']]
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
lg = LogisticRegression()
lg.fit(X_train,y_train)
```

```python
pred = lg.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
print('intercept',lg.intercept_)
print('coefficients',lg.coef_)
```

```python
df['boundary'] = (-lg.intercept_[0] - lg.coef_[0][0]*df['A'])/(lg.coef_[0][1])
```

```python
plt.xlim(0,80)
plt.ylim(0,90)
plt.scatter(x1,y1,c='red')
plt.scatter(x2,y2,c='blue')

plt.plot(df['A'],df['boundary'],ls='--',c='green')
```
