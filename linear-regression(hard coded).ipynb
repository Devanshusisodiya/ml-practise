## Defining some important functions

```python
learning_rate = 0.01
loss_arr = []
loss_arr_theta0, loss_arr_theta1 = [], []

# Functions to calculate the necessary calculations
# Hypothesis calculation
def h_theta(theta0, theta1, x):
    h = theta0 + theta1*x
    return h

# Loss calculation
def loss(hypo, true_val):
    err = (hypo - true_val)**2
    return err

# Gradient w.r.t. theta0 
def dloss_param0(hypo, y):
    partial_loss_0 = (hypo-y)
    return partial_loss_0

# Gradient w.r.t. theta1
def dloss_param1(hypo, x, y):
    partial_loss_1 = (hypo-y)*x
    return partial_loss_1
```

## The training process

```python
# Main Function including all operations
def main(param0, param1, train_feat, train_label, num, eta=learning_rate):
    # Training for 1000 epochs
    for _ in range(1000):
        # Looping over Training Data
        for i in range(len(train_feat)):
            
            # Calculating hypothesis and loss
            hypothesis = h_theta(param0, param1, train_feat[i])
            cur_loss = loss(hypothesis, train_feat[i])
            loss_arr.append(cur_loss/2*len(loss_arr))

            # Updating the parameters 
            temp0 = param0 - (eta*dloss_param0(hypothesis, train_label[i]))/num
            temp1 = param1 - (eta*dloss_param1(hypothesis, train_feat[i], train_label[i]))/num
            param0 = temp0
            param1 =  temp1
            loss_arr_theta0.append(param0)
            loss_arr_theta1.append(param1)
    return param0, param1
```

## The updated parameters

```python
# theta0 
main(10, 50, x, y, len(df))
```
